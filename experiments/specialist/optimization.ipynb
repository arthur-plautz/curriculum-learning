```python
from specialist_data import pipeline
from models.specialist.optimizer import Optimizer

manager = pipeline()
```

```python
scores = {}
params = {}

for seed, transformed in manager.transformed.items():
    optimizer = Optimizer()

    warm = transformed.data.query('index > 12000 and index < 12100')
    transformed.set_data(warm)

    optimizer.grid_search.fit(
        transformed.X_normalized,
        transformed.level
    )
    score_list = list(optimizer.grid_search.cv_results_.get('rank_test_score'))
    scores[seed] = max(score_list)
    index = score_list.index(max(score_list))
    params[seed] = optimizer.grid_search.cv_results_.get('params')[index]
```
