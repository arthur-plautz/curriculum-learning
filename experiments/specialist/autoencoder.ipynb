```python
from specialist_data import pipeline

manager = pipeline()
manager.set_target('s10')

X = manager.target_data.X
y = manager.target_data.y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras import Input, Model
from tensorflow.keras.layers import Dense, LeakyReLU, BatchNormalization

n_inputs = 6

def relu_and_layer(e, n):
    e = LeakyReLU()(e)
    e = Dense(n)(e)
    return e

def layer_and_relu(e, n):
    e = Dense(n)(e)
    e = LeakyReLU()(e)
    return e

i = Input(shape=(n_inputs,), name='initial_conditions')
e = BatchNormalization()(i)
e = relu_and_layer(e, 32)
e = relu_and_layer(e, 16)
e = relu_and_layer(e, 8)
e = Dense(3)(e)
e = layer_and_relu(e, 8)
e = layer_and_relu(e, 16)
e = layer_and_relu(e, 32)
e = BatchNormalization()(e)
o = Dense(6)(e)

model = Model(inputs=i, outputs=o, name='auto_encoder')

model.compile(optimizer='adam', loss='mse')
print(model.summary())
```

```python
history = model.fit(
    X_train, 
    X_train,
    epochs=200,
    batch_size=1000,
    verbose=2,
    validation_data=(X_test,X_test)
)
```
